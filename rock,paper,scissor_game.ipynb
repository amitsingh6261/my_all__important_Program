```python
import random
i=1
human_point=0
computer_point=0
i=0
while i<3:
 computer_choice=random.choice(["rock","paper","scissor"])
 human_choice=input("rock,paper,scissor me se ek select kar: ")

 if (human_choice=="rock" and computer_choice=="scissor") or (human_choice=="paper" and computer_choice=="rock") or (human_choice=="scissor" and computer_choice=="paper"):
   print("bhai tu jeet gya")
   human_point += 1
 elif human_choice==computer_choice:
   print("dono bhai dono tabhai")
 else:
   print("bhai tu machine se bhi har gya")
   computer_point += 1
 i=i+1

if human_point>computer_point:
  print(f"bhai tu series jeet gya {human_point}-{computer_point}")
elif human_point==computer_point:
  print("game draw hua {}-{}".formate(human_point,computer_point))
else:
  print(f"computer won tha game {human_point}-{computer_point}")

```
